```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dat = np.array([np.arange(1,11), np.arange(1,11)])
```

```python
plt.scatter(dat[0], dat[1], c='r')
plt.plot(dat[0], dat[1])
plt.xlabel('number of bedrooms')
plt.ylabel('price of house in millions')
plt.show()
```
